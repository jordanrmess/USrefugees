```python
%%capture
!pip install folium

import folium

import requests

import pandas
```

```python
arrest_table = pandas.read_csv("http://www.hcbravo.org/IntroDataSci/misc/BPD_Arrests.csv")
```

```python
arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]
```

```python
arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str

arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)

arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)
```

```python
arrest_table.head()
```

```python
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster
import random
```

```python
heat_data = [[row['lat'], row['long']] for index,row in arrest_table.iterrows()]
```

This map shows the heat map for a random sample of 1000 of the crimes. It shows us
what areas of Baltimore have the most crime. This is useful data because we can see that
as expected, there is a lot more crime in the inner city part of Baltimore, and as we 
get further from inner city, there is less crime. However, there are still parts of the
suburbs that have a lot of crime, as indicated by the map. This map is interactive because
as you zoom in and out, the heat map adjusts to be more accurate of the exact locations
you are looking at.

```python
map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)
HeatMap(random.sample(heat_data, 1000)).add_to(map_osm)
map_osm
```

This map shows the clustered map of narcotics crimes for men in the Baltimore area. We used a
random sample of 1000 crimes. This map is interactive because as you zoom in, the clusters
readjust to be more accurate of the area you are looking at.

```python
map1 = folium.Map(location=[39.29, -76.61], zoom_start=11)
df = arrest_table
narcotics_male = df[(df.incidentOffense == '87-Narcotics') & (df.sex == 'M')].sample(n=1000)

marker_cluster = MarkerCluster().add_to(map1)
for index, row in narcotics_male.iterrows():
    folium.Marker(location=[row['lat'],row['long']]).add_to(marker_cluster)
                                                                                             
map1
```

This map shows the clustered map of narcotics crimes for women in the Baltimore area. We used a random sample of 1000 crimes. This is the same technique as the previous map, and is
also interactive.

```python
map2 = folium.Map(location=[39.29, -76.61], zoom_start=11)

df = arrest_table
narcotics_female = df[(df.incidentOffense == '87-Narcotics') & (df.sex == 'F')].sample(n=1000)

marker_cluster = MarkerCluster().add_to(map2)
for index, row in narcotics_female.iterrows():
    folium.Marker(location=[row['lat'],row['long']]).add_to(marker_cluster)
                                                                                             
map2
```

Looking at the maps for narcotic arrests for men and women, we can see that they are fairly 
equal in occurrences and they occur in the same areas. This suggests that gender does not
give us insight into whether or not the crime will be narcotics related, because they have
equal occurrences of happening.

The following map shows the aggravated assault occurrences for men. We used a random sample of 300.

```python
map3 = folium.Map(location=[39.29, -76.61], zoom_start=11)

df = arrest_table
agg_male = df[(df.incidentOffense == '4B-Agg. Asslt.- Cut') & (df.sex == 'M')].sample(n=300)

marker_cluster = MarkerCluster().add_to(map3)
for index, row in agg_male.iterrows():
    folium.Marker(location=[row['lat'],row['long']]).add_to(marker_cluster)
                                                                                             
map3
```

The following map shows aggravated assault occurrences for women, also using a sample size
of 300.

```python
map4 = folium.Map(location=[39.29, -76.61], zoom_start=11)

df = arrest_table
agg_female = df[(df.incidentOffense == '4B-Agg. Asslt.- Cut') & (df.sex == 'F')].sample(n=300)

marker_cluster = MarkerCluster().add_to(map4)
for index, row in agg_female.iterrows():
    folium.Marker(location=[row['lat'],row['long']]).add_to(marker_cluster)
                                                                                             
map4
```

Based on the two above maps, we can see that there are much fewer occurrences of aggravated
assault in women. We had to use a smaller sample size because there were not enough data points
to get as many male occurrences. We can also see that we had to use smaller sample sizes
for both men and women because there are not as many aggravated assault charges in Baltimore
as there are narcotics. These maps can give us insight into the correlation between not only
gender and crime, but also location and crime. In regards to the gender, we can see that men
have higher reported aggravated assault charges. In regards to location, we can see that 
from the data, the aggravated assault numbers are much lower than the narcotics numbers.
